```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

## Question1 Part1

```python
f1 = open("karate/zachary.dat", "r")
df_karate = f1.readlines()
df_zachestr = np.array(df_karate[7:41])
# df_zachcstr = np.array(df_karate[41:])

df_zache = np.empty((0, len(df_zachestr)), int)

for i in df_zachestr:
    df_zache = np.append(df_zache, [np.fromstring(i, dtype=int, sep=' ')],  axis=0)  
```

```python
G_zache = nx.from_numpy_matrix(df_zache)
nx.draw(G_zache)
```

```python
print("Zache is connected. ", nx.is_connected(G_zache))
max_degree = 0
sorted(G_zache.degree, key=lambda x: x[1], reverse=True)
print("Max degree for Zache is ", G_zache.degree[0])
print("Diameter of Zache is", nx.diameter(G_zache))
print("Clustering Coefficient of Zache is", nx.average_clustering(G_zache))
```

```python
spl_zache = dict(nx.shortest_path_length(G_zache))
pair_zache = list(combinations(G_zache.nodes(), 2))
pairvalue_zache = []
for pair in pair_zache:
    pairvalue_zache = np.append(pairvalue_zache, spl_zache[pair[0]][pair[1]])


pairvalue_zache = list(pairvalue_zache)
pairvalue_zache = [int(i) for i in pairvalue_zache]

# print(pairvalue_zache)
fig, ax = plt.subplots()
hist = ax.hist(pairvalue_zache)
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
```

```python
# print(G_zache.degree())
list_node, list_degree = zip(*G_zache.degree())
fig, ax = plt.subplots()
hist = ax.hist(list_degree)
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
```

```python
bc_zache = nx.betweenness_centrality(G_zache)
print(bc_zache)
bcvalue_zache = list(bc_zache.values())
plt.hist(bcvalue_zache)
plt.show()
```

```python
local_cc = nx.clustering(G_zache)
plt.hist(list(local_cc.values()))
plt.show()
```

```python
f_grouplist = open("karate/groups.txt", "r")
fin = f_grouplist.readlines()
group_list = []
for group in fin:
    group_list.append(group.replace('\n', "") )

df_zache = nx.to_pandas_adjacency(G_zache)
df_zache.index = group_list
df_zache.columns = group_list
# print(df_zache)
df_zache.to_csv("p1.csv")
```

## Question1 part3

```python
# data load and preprocess

f_prison = open("prison/prison.dat", "r")
fin_prison = f_prison.readlines()
df_prisonstr = np.array(fin_prison[4:71])
# print(df_prisonstr)

df_prison = np.empty((0, len(df_prisonstr)), int)
for i in df_prisonstr:
    df_prison = np.append(df_prison, [np.fromstring(i, dtype=int, sep=' ')],  axis=0)
df_prison = pd.DataFrame(df_prison)
df_prison.to_csv("p3.csv")
```

```python
G_prison = nx.from_pandas_adjacency(df_prison,create_using=nx.DiGraph())
nx.draw(G_prison)
```

```python
print("Prison is connected. ", nx.is_strongly_connected(G_prison))
max_degree = 0
sorted(G_prison.degree, key=lambda x: x[1], reverse=True)
print("Max degree for Prison is ", G_prison.degree[0])
# print("Diameter of Dophin is", nx.diameter(G_prison))
print("Clustering Coefficient of Prison is", nx.average_clustering(G_prison))
```

```python
# Degree of each node
list_node, list_degree = zip(*G_prison.degree())
fig, ax = plt.subplots()
hist = ax.hist(list_degree)
ax.xaxis.set_major_locator(plt.MultipleLocator(1))
```

```python
# betweeness centrality
bc_prison = nx.betweenness_centrality(G_prison)
print(bc_prison)
bcvalue_prison = list(bc_prison.values())
plt.hist(bcvalue_prison)
plt.show()
```

```python
# local clustering coefficient
local_cc = nx.clustering(G_prison)
plt.hist(list(local_cc.values()))
plt.show()
```
